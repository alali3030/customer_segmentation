# Customer Segmentaion Analysis

This notebook demonstrates multiple customer segmentation techniques using online shopping data

The full dataset can be found here: https://archive.ics.uci.edu/ml/datasets/Online+Retail+II

```python
#import required libraries
import pandas as pd
import datetime as dt
import seaborn as sns
sns.set()
```

```python
#upload a random sample of 20% from the original dataset
online = pd.read_csv('online.csv')
online = online.iloc[:,1:]
#convert date from string to a datetime object
online['InvoiceDate'] = online['InvoiceDate'].apply(lambda x: dt.datetime.strptime(x, '%Y-%m-%d %H:%M:%S'))
```

```python
#The dataset shape & head
print(online.shape)
online.head()
```

```python
#Calculate the total price of an invoice
online['TotalPrice'] = online['Quantity'] * online['UnitPrice']
```

```python
#creat a basket ID that can be used for aggregating customer baskets based on customer & invoice IDs
online['BasketID'] = online['CustomerID'].astype(str) + online['InvoiceDate'].astype(str)
```

```python
#create additional columns for invoice and cohort months
#first create a function that returns the first day of a month
def get_month(x): return dt.datetime(x.year, x.month, 1)

online['InvoiceMonth'] = online['InvoiceDate'].apply(get_month)
grouping = online.groupby('CustomerID')['InvoiceMonth']
online['CohortMonth'] = grouping.transform('min')

online.head()
```

```python
#create a helper function to get the year, month, and day from the date
def get_date_int(df, column):
    year = df[column].dt.year
    month = df[column].dt.month
    day = df[column].dt.day
    
    return year, month, day
```

```python
#assign time offset between cohort and invoice dates

invoice_year, invoice_month, _ = get_date_int(online, 'InvoiceMonth')
cohort_year, cohort_month, _ = get_date_int(online, 'CohortMonth')
years_diff = invoice_year - cohort_year
months_diff = invoice_month - cohort_month

online['CohortIndex'] = years_diff * 12 + months_diff + 1 #added 1 to avoid 0 so joining month is 1
online.head()
```

#### Here we create the cohort pivot table to be able to calculate some metrics

```python
#counth the monthly active customers from each cohort
grouping = online.groupby(['CohortMonth', 'CohortIndex'])
cohort_data = grouping['CustomerID'].apply(pd.Series.nunique)
cohort_data = cohort_data.reset_index()
cohort_counts = cohort_data.pivot(index='CohortMonth',
                                  columns='CohortIndex',
                                  values='CustomerID')

cohort_counts
```

## Cohort Metrics

```python
#calculate the retention rate
#first month is 100% by default
cohort_sizes = cohort_counts.iloc[:,0]
retention = cohort_counts.divide(cohort_sizes, axis=0)
retention.round(3)*100
```

```python
#calculate hte average quantity for each cohort month
grouping = online.groupby(['CohortMonth', 'CohortIndex'])
cohort_data = grouping['Quantity'].mean()
cohort_data = cohort_data.reset_index()
average_quantity = cohort_data.pivot(index='CohortMonth',
                                    columns='CohortIndex',
                                    values='Quantity')

average_quantity.round(1)
```

## Recency, Frequency, Monitary Value Segmentation (RFM) 

There are main three indicators that differentiates valuable customers from others:

- Recency(R) when was the last purchase as an evidence that the customer is still active
- Frequency (F) how often does the customer buy to demonstrate loyalty
- Monetary Value (M) how much revenue is generated from the customer to capture the gained value

Two more indicators are added here as an experiment:

- Volume (V) What is the basket size or quatity of a purchase
- Tenure (T) How long has the customer been retained aka retention

These point together helps in highlighting the purchasing behaviour of customers and enables optimizing and upselling various segments

```python
#use the last day as the most recent day (should be the day of the analysis in pactice)
snapshot_date = max(online['InvoiceDate']) + dt.timedelta(days=1)
```

```python
#Create a column for days since invoice to be used for tenure
today = max(online['InvoiceDate'])
online['days'] = online['InvoiceDate'].apply(lambda x: (today - x).days +1)
```

```python
online.head()
```

```python
#Used for tableau viz
#online.to_csv('online_full.csv')
```

```python
#aggregate the data to customer level with RFM
datamart = online.groupby(['CustomerID']).agg({
    'InvoiceDate': lambda x: (snapshot_date - x.max()).days,
    'InvoiceNo': 'count',
    'TotalPrice': 'sum',
    'Quantity' : 'max',
    'days' : 'max'})

#rename columns to RFM for convenience
datamart.columns = ['Recency', 'Frequency', 'MonetaryValue', 'Volume', 'Tenure']

datamart.head()
```

### Building RFM Segmentation

Now we have the three metrics for each CustomerID that can be used for segmentation and further analysis

```python
#segment rcency by quartiles to enable comparative analysis
r_labels = range(4, 0, -1)
r_quartiles = pd.qcut(datamart['Recency'], 4, labels = r_labels)
datamart = datamart.assign(R = r_quartiles.values)
```

```python
#do the same for the rest
f_labels = range(1,5)
m_labels = range(1,5)
v_labels = range(1,5)
t_labels = range(1,5)

f_quartiles = pd.qcut(datamart['Frequency'], 4, labels = f_labels)
m_quartiles = pd.qcut(datamart['MonetaryValue'], 4, labels = m_labels)
v_quartiles = pd.qcut(datamart['Volume'], 4, labels = v_labels)
t_quartiles = pd.qcut(datamart['Tenure'], 4, labels = t_labels)

datamart = datamart.assign(F = f_quartiles.values)
datamart = datamart.assign(M = m_quartiles.values)
datamart = datamart.assign(V = v_quartiles.values)
datamart = datamart.assign(T = t_quartiles.values)

datamart.head()
```

```python
#the next step is to give each customer an RFM score which can be based on the three values
def join_rfm(x): return str(x['R']) + str(x['F']) + str(x['M']) + str(x['V']) + str(x['T'])

datamart['RFM_Segment'] = datamart.apply(join_rfm, axis=1)
datamart['RFM_Score'] = datamart[['R','F','M', 'V', 'T']].sum(axis=1)
```

```python
#grouping customers into named segments for clarity of insights
def segment_me(df):
    if df['RFM_Score'] >= 15:
        return 'Gold'
    elif (df['RFM_Score'] >= 10) and (df['RFM_Score'] < 15):
        return 'Silver'
    else:
        return 'Bronze'


datamart['General_Segment'] = datamart.apply(segment_me, axis=1)
datamart.groupby('General_Segment').agg({'Recency': 'mean'
,'Frequency': 'mean'
,'MonetaryValue': 'mean'
,'Volume' : 'mean'
,'Tenure' : ['mean'
,'count']}).round(1)
```

## K-means clustering

```python
#k-means clustering expects an equal mean and std between veriables 
#so preprocessing is required to standardize RFM
from sklearn.preprocessing import StandardScaler
import numpy as np
#log transformation
datamart_rfm = datamart.iloc[:,:5].copy()
datamart_log = np.log(datamart_rfm)
#scaling
scaler = StandardScaler()
scaler.fit(datamart_log)
datamart_normalized = scaler.transform(datamart_log)

print('mean: ', datamart_normalized.mean(axis=0).round(2))
print('std: ', datamart_normalized.std(axis=0).round(2))
```

```python
#we use the eblow method to select the optimal cluster number
# Import key libraries
from sklearn.cluster import KMeans
import seaborn as sns
from matplotlib import pyplot as plt

# Fit KMeans and calculate SSE for each *k*
sse = {}
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=1)
    kmeans.fit(datamart_normalized)
    sse[k] = kmeans.inertia_
    
# sum of squared distances to closest cluster center
# Plot SSE for each *k*
plt.title('The Elbow Method')
plt.xlabel('k'); plt.ylabel('SSE')
sns.pointplot(x=list(sse.keys()), y=list(sse.values()))
plt.show()
```

```python
#4 clusters is selected
kmeans = KMeans(n_clusters=4, random_state=1)
kmeans.fit(datamart_normalized)
cluster_labels = kmeans.labels_

#add cluster label to original dataset
datamart_rfm_k4 = datamart_rfm.assign(Cluster = cluster_labels)
```

```python
#Clustering function with scaling and log trx
from sklearn.cluster import KMeans
def cluster(dataset, n):
    from sklearn.cluster import KMeans
    scaler = StandardScaler()
    data_log = np.log(dataset)
    scaler.fit(data_log)
    data_log_norm = scaler.transform(data_log)
    kmeans = KMeans(n_clusters=n, random_state=1)

    kmeans.fit(data_log_norm)

    cluster_labels = kmeans.labels_


    data_log_norm = pd.DataFrame(data_log_norm.round(2),
                                     index = dataset.index,
                                     columns = dataset.columns)
    gr4_kn = data_log_norm.assign(Cluster = cluster_labels)
    return gr4_kn
```

```python
#use the clustering function based on eblow value
x = cluster(datamart.iloc[:,:5].copy(), 4)
```

## Dimensionality reduction and then clustering

The data has 5 variables so not highly dimensional. However, it helps to visualize the data on a 2d scatter plot

```python
#use TSNE to reduce dimensionality to 2d using the normalized data without passing the cluster value

from sklearn.manifold import TSNE

tsne_embedded = TSNE(n_components=2, random_state=42, perplexity = 100, learning_rate=50, n_iter=5000).fit_transform(x.drop('Cluster', axis = 1))
```

```python
#plot the clustering map

plt.figure(figsize=(10, 7))
sns.scatterplot(data = pd.DataFrame(tsne_embedded), x=0, y=1)
plt.title('Clustering Customer Segments')
plt.show()
```

```python
#Assess number of clusters on lower dimensional data
sse = {}
for k in range(1, 16):
    kmeans = KMeans(n_clusters=k, random_state=1)
    kmeans.fit(tsne_embedded)
    sse[k] = kmeans.inertia_

# Plot SSE for each *k*
plt.title('The Elbow Method')
plt.xlabel('k'); plt.ylabel('SSE') 
sns.pointplot(x=list(sse.keys()), y=list(sse.values()))
plt.show()
```

```python
#Cluster data on 2d
df_tsne = pd.DataFrame(tsne_embedded)
kmeans = KMeans(n_clusters=4, random_state=1)

kmeans.fit(pd.DataFrame(df_tsne))

cluster_labels = kmeans.labels_


df_tsne = pd.DataFrame(df_tsne,
                                 index = df_tsne.index,
                                 columns = df_tsne.columns).round(2)
gr4_kn = df_tsne.assign(Cluster = cluster_labels)
```
